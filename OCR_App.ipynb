```python
import matplotlib.pyplot as plt
import keras_ocr
```

```python
images = [
          plt.imread("/content/1.jpg"), 
          plt.imread("/content/2.jpg")
]
```

```python
#keras-ocr will automatically download pretrained weights for the detector and recognizer.
pipeline = keras_ocr.pipeline.Pipeline()

#  Each list of predictions is a list of(word, box) tuples.
predictions = pipeline.recognize(images)
```

```python
# creating seperate plots for each image
fig, axs = plt.subplots(nrows=len(images), figsize=(8,8))

for ax, image in zip(axs, images):
  ax.axis('off')
  ax.imshow(image)
```

```python
# creating seperate plots for each image
fig, axs = plt.subplots(nrows=len(images), figsize=(14,14))

for ax, image, predict in zip(axs, images, predictions):
  keras_ocr.tools.drawAnnotations(image=image, predictions=predict, ax=ax)
```
